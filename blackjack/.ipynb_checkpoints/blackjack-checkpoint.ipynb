```python
#face cards 10
#Aces value 1
#Bankroll
#deckshuffle
#playerhand
#computerhand
#win p>c
#win dealer bust
#lose p bust
#win = bet *2
from random import shuffle
card_values={'ace':1,'two':2,'three':3,'four':4,'five':5,'six':6,'seven':7,'eight':8,'nine':9,'ten':10,'jack':10,'queen':10,'king':10}
card_suit=('hearts','diamonds','spades','clubs')
card_rank=('two','three','four','five','six','seven','eight','nine','ten','jack','queen','king','ace')
```

```python
class Card:
    def __init__(self, rank,suit):
        self.rank=rank
        self.suit=suit
        self.value=card_values[rank]
    def __str__(self):
        return f'{self.rank} of {self.suit}  '
```

```python
class Deck:
    def __init__(self):
        self.all_cards=[]
        for i in card_suit:
            for j in card_rank:
                self.all_cards.append(Card(j,i))
    
    def shuf(self):
        shuffle(self.all_cards)
    
    def deal1(self):
        return self.all_cards.pop()
    
   
                
```

```python
class Player:
    def __init__(self,chips,name):
        self.chips=chips
        self.hand=[]
        self.sum_hand=0
        self.name=name
        
    def add_cards(self,c):
        self.hand.append(c)
        
    def val_of_hand(self):
        self.sum_hand=0
        for i in self.hand:
            
            self.sum_hand+=i.value
        return self.sum_hand
    def clear_hand(self):
        self.hand=[]
    
```

```python
game_on=True
plr = Player(100,'Player1')
dlr = Player(0,'Dealer')

def check_bust(entity):
    if entity.val_of_hand()>=21:
        return True
    return False
    

while game_on:
    plr.clear_hand()
    dlr.clear_hand()
    dck = Deck()
    dck.shuf()
    for i in range(2):
        plr.add_cards(dck.deal1())
        dlr.add_cards(dck.deal1())
    player_playing=True
    dealer_playing=True
    
    
    if plr.chips==0:
        print('Player has insufficient funds to play')
        break
    
    print('=====================')
    print(f'Available funds ${plr.chips}')
    print('=====================')
    bet=int(input('Enter your bet'))
    
    while player_playing:
        print('=====================')
        print("Dealer Hand")
        for i in range(len(dlr.hand)-1):
            print(dlr.hand[i],end=' ')
            print('XXX')
        print("Player Hand") 
        for i in plr.hand:
            print(i, end=' ')
        print(f'Value of player hand {plr.val_of_hand()}')
        print('')
        print('=====================')
        print('Enter 1 to Hit')
        print('Enter 2 to stay')
        x=int(input("Enter: "))
        if x==1:
            plr.add_cards(dck.deal1())
        elif x==2:
            player_playing=False
        else:
            print('Invalid Input Try Again')
        if check_bust(plr):
            break
            
    while dealer_playing:
        print('=====================')##############
        for i in dlr.hand:
            print(i,end=' ')
        print(f'Value of dealer hand {dlr.val_of_hand()}')
        for i in plr.hand:
            print(i, end=' ')
        print(f'Value of player hand {plr.val_of_hand()}')
        print('=====================')
        #Checks busts
        if check_bust(plr):
            print('Player Bust! Dealer Win')
            plr.chips-=bet
            break
        if check_bust(dlr):
            print("Dealer bust! Player Wins")
            plr.chips+=(bet)
            dealer_playing=False
        #Automated dealer logic
        elif dlr.val_of_hand()>plr.val_of_hand():
            plr.chips-=bet
            print('Dealer Wins')
            dealer_playing=False
            #stand
        elif dlr.val_of_hand()==plr.val_of_hand() and dlr.val_of_hand()>=17:
            print('Draw!')
            dealer_playing=False
        else:
            dlr.add_cards(dck.deal1())
            
    if input("Quit?: y/n ")=='y':
        break

            
        
        
        
        
            
        
    
```
